```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_dir = "/Users/iben/Desktop/MBML/Project_MBML/data/Smartphone_Usage_And_Addiction_Analysis_7500_Rows (1).csv"

df = pd.read_csv(data_dir)

# Display the first few rows of the DataFrame
#print(df.head())

# Check for missing values
print(df.isna().sum())

# Check for none values
print(df.isnull().sum())

# print academic performance column
print(df['academic_work_impact'].value_counts())
```

```python
# replace NaN values with 'None' in the 'addiction_level' column
df['addiction_level'] = df['addiction_level'].fillna('None')


#df['hours_spent_on_smartphone_per_day'] = df['hours_spent_on_smartphone_per_day'].fillna(mean_hours)

# plot histogram of addiction levels
plt.figure(figsize=(10, 6))
plt.hist(df['addiction_level'], bins=4, color='skyblue', edgecolor='black')
plt.title('Distribution of Smartphone Addiction Levels')
plt.xlabel('Addiction Level')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# gender distribution
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()


# min age
min_age = df['age'].min()
# max age
max_age = df['age'].max()
num_bins = max_age - min_age + 1

# age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=num_bins, color='lightcoral', edgecolor='black')
plt.title('Age Distribution of Participants')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# plot scatter plot of hours spent on smartphone per day vs addiction level
plt.figure(figsize=(10, 6))
plt.scatter(df['daily_screen_time_hours'], df['addiction_level'], color='lightgreen', edgecolor='black')
plt.title('Daily Screen Time Hours vs Addiction Level')
plt.xlabel('Daily Screen Time Hours')
plt.ylabel('Addiction Level')
plt.grid()


 

```
